```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

replacing blanks with 0 as tenure is 0 and no total charges

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"
    
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
df.head(30)
```

converted 0 and 1 values senior citizen to yes/no

```python
ax = sns.countplot(x = df['Churn'], data = df)
ax.bar_label(ax.containers[0])
plt.title("Count customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})

plt.title("Churned customers percentage", fontsize ="9")
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.show()
```

From the given piechart we can conclude 26.54% customers are churned out

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender", data = df, hue ='Churn')
plot_color = 'orange'
plt.title('Churn by Gender', fontsize ="9")
plt.show()
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins = 72, hue = "Churn")
plt.show()
```

People who used our services for long time have stay

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue ='Churn')
ax.bar_label(ax.containers[0])
plt.title('Customers Count by Contract', fontsize =10)
plt.show()
```

People who have month to month plan are likely to churn out than have longer contract

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']

num_columns = 3 
num_rows = (len(columns) // num_columns) + (len(columns) % num_columns)

fig, axes = plt.subplots(num_rows, num_columns, figsize=(15, 10))
axes = axes.flatten()
#plot_color = 'orange'  # You can change this to any color you'd like

# Loop over each column and create a countplot
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], color=plot_color, hue =df['Churn'])
    axes[i].set_title(f'Countplot of {col}', fontsize=14)
    
    #Increase font size for x-axis labels and keep them horizontal
    axes[i].tick_params(axis='x', labelsize=12)
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=0, ha='center')  # Keep labels horizontal

# Remove any empty subplots if the number of plots is not a perfect grid
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])
plt.tight_layout()
plt.show()
```

The majority customers who do not churn tend to have services like Phone Service , Internet service(particularly DSL) and Online security enabled. For this services like Onlinke backup, Tech Support and Streaming Tv churn rates are noticebly higher when these servces are not used or unavailabel.
